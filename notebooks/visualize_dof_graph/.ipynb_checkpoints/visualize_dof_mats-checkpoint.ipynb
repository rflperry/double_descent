```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from scipy.spatial.distance import pdist, squareform
import pickle

import sys
sys.path.append('../../functions/')
from dataset import generate_gaussian_parity
```

```python
X_train, y_train = generate_gaussian_parity(n=1000, angle_params=0, acorn=1234)
```

```python
widths =  [4, 8, 16, 32]
depths = [1, 2, 3, 4]

for depth in depths:
    for width in widths:
        with open(f'./activation_matrices/xor_results_depth={depth}_width={width}.pkl', 'rb') as f:
            results_dict = pickle.load(f)
        
        act_mat = np.asarray(results_dict['activations'])#[:, :width]
        unique_acts, counts = np.unique(act_mat, axis=0, return_counts=True)
        kernel_mat = 1 - squareform(pdist(unique_acts, p=1)) / unique_acts.shape[1]
        region_posteriors = []
        for act in unique_acts:
            idx = np.all(act_mat == np.asarray(act), axis=1)
            region_posteriors.append(np.mean(y_train[idx]))
        G = nx.from_numpy_matrix(kernel_mat)
        nx.draw_spring(G, node_color=region_posteriors, node_size=counts, cmap=plt.cm.RdBu_r, vmin=0, vmax=1)
        plt.title(f"Model depth={depth}, width={width}, train_acc={results_dict['train_acc']}, test_acc={results_dict['test_acc']}")
        sm = plt.cm.ScalarMappable(cmap=plt.cm.RdBu_r, norm=plt.Normalize(vmin = 0, vmax=1))
        plt.colorbar(sm)
        plt.show()
```
